```python
#Import modules
import pyodbc
import pandas as pd
```

```python
#Connect to DB
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=jvsql02;'
                      'Database=DIVE;'
                      'Trusted_Connection=yes;')

#Select the data from the table
SQL_Query = pd.read_sql_query(
'''select * from DIVE.dbo.TCR_OGSYS_CAPEX_GENLEDGER''', conn)
```

```python
#Convert the query to a data frame
df = pd.DataFrame(SQL_Query, columns=              
    [
        'Company_Code',  'Main_Account'
        , 'Sub_Account', 'Location_Type', 'Location_Code', 'Journal_Date', 'Code'
        ,'Posted', 'Include_in_Journal_Report', 'Present_in_Journal_Balance'
        , 'Accrual_Date', 'Include_in_Accrual_Report', 'Present_in_Accrual_Balance'
        , 'Cash_Date', 'Include_in_Cash_Report', 'Present_in_Cash_Balance'
        , 'Voucher_Code', 'Expense_Deck_Source_Serial', 'Revenue_Deck_Source_Serial'
        , 'Source_Module_Code', 'Description', 'Reference_Type', 'Reference', 'Entity_Type'
        , 'Entity_Code', 'Well_Code', 'AFE_Code', 'Expense_Deck_Code'
        , 'Expense_Deck_Version_Date', 'Expense_Deck_Change_Code', 'Expense_Deck_Interest_Total'
        , 'Reconciliation_Type', 'Reconciled_Trial', 'Reconciled', 'Generated_Entry'
        , 'Voucher_Uniqueness_ID', 'Gross_88ths_Volume'
        , 'Net_Volume', 'Gross_88ths_Value', 'Net_Value', 'Journal_Fiscal_Year'
        , 'Cash_Fiscal_Year', 'Accrual_Fiscal_Year', 'Entry_Group'
        , 'Entry_Sequence', 'Posting_Thread_ID', 'Posting_Task_ID', 'Cash_Date_Frozen'
        , 'Well_Allocation_Deck_Type', 'Well_Allocation_Deck_Code'
        , 'Well_Allocation_Product_Code', 'Well_Allocation_Purpose', 'Well_Allocation_Version_Date'
        , 'Well_Allocation_ID', 'Distribution_Company_Password_Accepted'
        , 'AccountNumber', 'AccountName'
    ])
```
