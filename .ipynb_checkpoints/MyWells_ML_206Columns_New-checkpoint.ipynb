```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
```

```python
sample.describe()
```

```python
sample.shape
```

```python
sample.dropna().shape
```

```python
dt = DecisionTreeClassifier()
```

```python
features = sample.dropna().drop(label_column, axis=1)
labels = sample.dropna()[label_column]

X = features.values
y = labels.values
```

```python
dt.fit(X, y)
```

```python
feature_importances = dt.feature_importances_
feature_importances
```

```python
import numpy as np
```

```python
# Get non zero indices to filter with
nz_indices = np.argwhere(feature_importances > 0)
nz_indices.shape
```

```python
nz_features = features.columns[nz_indices].values.flatten()
nz_features
```

```python
nz_importances = feature_importances[nz_indices].flatten()
nz_importances
```

```python
plt.xticks(rotation=90)
plt.bar(nz_features, nz_importances, log=True)
```

```python
sorted_indices = np.argsort(nz_importances)
```

```python
sorted_features = nz_features[sorted_indices]
```

```python
sorted_importances = nz_importances[sorted_indices]
```

```python
plt.xticks(rotation=90)
plt.bar(sorted_features, sorted_importances, log=True)
```
